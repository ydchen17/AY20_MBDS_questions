Given a L by L square grid, consider 4 neighbors connections. Given L 2 beads of different colours. Put all the beads onto the grid and the penalty for putting any two beads of the same colour as neighbor is one. Total penalty is the sum of all panelties. Your task is to Ô¨Ånd a way to put the beads onto the square grid with least penalty. Perform this tasks with

1. L=5 with 12 red beads (R) and 13 blue beads (B)
2. L=64 with 139 red beads (R), 1451 blue beads (B), 977 green beads (G), 1072 white beads (W), 457 yellow beads (Y)

**Question 5.1**

To get the lowest penalty, beads with the same colour should be seperated by those with other colours. Consider only two colours with the same amount of beads, the pattern would be as below, which produces no penalty.

R B R B R <br>
B R B R B <br>
R B R B R <br>
B R B R B <br>

**Question 5.2**

Ideally, we can classify all beads into two groups, of each which do not share the colours. Furthermore, we expect that the number of beads in two categories should be highly alike but not the same.

Let's consider two categories, A and B. Consider the complexity to calculate a global optimality, we will break up the question into smaller questions, like the grid that we are familar with right now. 

Hence, when we fill the grid, we will start from up left side to the down right. In each diagonal line, we will mix A with B, unless, let's say, B is used up, and there are R-B grids remaining. Then, we just reduce the number in category A by (A-B) and then repeatedly trying to make the number of A and B close to each other. 


```python
L=64
output = [['']*L for _ in range(L)]    
freqlist=[139,457,977,1072,1451]
letterlist=['R','Y','G','W','B']
bestchoose=[0,0,1,1,0]
freqlist[3]-=1 

for i in range(0,L*2-1):
    colourwhom=(i%2+1)%2
    for t in range(0,(L-1)-abs(i-(L-1))+1):
        for j in range(0,len(bestchoose)):
            if bestchoose[j]==colourwhom and freqlist[j]>0:

                thisroundcolour=letterlist[j]
                if (i<=63):
                    output[i-t][t]=thisroundcolour
                else:
                    output[i-t-(i-L+1)][t+(i-L+1)]=thisroundcolour
                freqlist[j]-=1
                break
                
for i in range(0,L):
    for j in range(0,L):
        if (output[i][j]==''):
            output[i][j]='G'

for row in output:
    combine = ""
    for letter in row:
        combine += letter+" "
    #print (combine)
```
