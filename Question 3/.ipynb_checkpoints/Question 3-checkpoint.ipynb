```python
# Import our dataset
# Here I follow the guideline at https://www.tensorflow.org/tutorials/load_data/text
import tensorflow_datasets as tfds
import os

# Let's see where I am and whether we have the files we need locally, as I am running this remotely. 
print (os.getcwd())
print (os.listdir(os.getcwd()))
```

```python
from sklearn.neural_network import MLPRegressor
import numpy as np

#Loading all data
train_data = np.loadtxt("train_data.txt", skiprows=1)
train_truth = np.loadtxt("train_truth.txt", skiprows=1)
test_data = np.loadtxt("test_data.txt", skiprows=1)

# Training with MLP
clf = MLPRegressor(hidden_layer_sizes=(4,4))
clf.fit(train_data,train_truth)

# Prediction
predict_truth = clf.predict(test_data)

# Writing the output
output= open("test_predict.txt","a")
output.write("y\n")
for i in predict_truth:
    output.write(str(i))
    output.write("\n")
output.close()
```
