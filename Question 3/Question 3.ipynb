```python
from sklearn.neural_network import MLPRegressor
import numpy as np

#Loading all data
train_data = np.loadtxt("train_data.txt", skiprows=1)
train_truth = np.loadtxt("train_truth.txt", skiprows=1)
test_data = np.loadtxt("test_data.txt", skiprows=1)

# Training with MLP
clf = MLPRegressor(hidden_layer_sizes=(4,4))
clf.fit(train_data,train_truth)

# Prediction
predict_truth = clf.predict(test_data)

# Writing the output
output= open("test_predict.txt","w")
output.write("y\n")
for i in predict_truth:
    output.write(str(i))
    output.write("\n")
output.close()
```
